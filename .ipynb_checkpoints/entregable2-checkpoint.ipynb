# Ejercicio 1 SQL:

1. Crear una base de datos en SQLite utilizando la libreria SQLalchemy. https://docs.sqlalchemy.org/en/14/core/engines.html#sqlite

2. Ingestar los datos provistos en 'https://cs.famaf.unc.edu.ar/~mteruel/datasets/diplodatos/melb_data.csv' en una tabla y el dataset generado en clase con datos de airbnb y sus precios por codigo postal en otra.

3. Implementar consultas en SQL que respondan con la siguiente información:

- cantidad de registros totales por ciudad.
- cantidad de registros totales por barrio y ciudad.

4. Combinar los datasets de ambas tablas ingestadas utilizando el comando JOIN de SQL para obtener un resultado similar a lo realizado con Pandas en clase.

```python
# Create the database connection using SQLalchemy

engine = create_engine(
    r"{}:///{}".format(
        dbENGINE,
        os.path.join(
            os.getcwd(),
            dbNAME,
        )
    ), 
    echo=True
)
```

```python
# Ingest records into the database
kaggle_df.to_sql("kaggle_melbourne", con=engine, if_exists="replace", index=False)
```

# Ahora leemos el dataset de Airbnb que contiene precios por código postal

```python
airbnb_df_zipcode = pd.read_csv(os.path.join(os.getcwd(), "airbnb_price_by_zipcode.csv"))
```

```python
airbnb_df_zipcode
```

```python
airbnb_df_zipcode.to_sql('zip_airbnb_melbourne', con=engine, if_exists="replace")
```

# SQL queries using SQLalchemy

```python
# Cantidad de registros totales por ciudad en dataset Kaggle
query1 = "SELECT CouncilArea, COUNT(CouncilArea) AS Cantidad FROM kaggle_melbourne GROUP BY CouncilArea"
          
# Cantidad de registros totales por barrio y ciudad en Kaggle
query2 = "SELECT CouncilArea, Suburb, COUNT(Suburb) AS Cantidad FROM kaggle_melbourne GROUP BY Suburb ORDER BY CouncilArea"

# Listamos los queries de consulta
queries = [query1, query2]

# Conectamos a la base de datos y sistematizamos las consultas con un bucle For
with engine.connect() as con:
    for i, query in enumerate(queries):
        rs = con.execute(query)
        print(query)
        print("\n")
        if i == 0:
            print("SE IMPRIME LA CANTIDAD DE REGISTROS TOTALES POR CIUDAD")
        else:
            print("SE IMPRIME LA CANTIDAD DE REGISTROS TOTALES POR BARRIO Y CIUDAD")
        print("\n")
        for row in rs:
            if i == 0:
                print(f"Ciudad: {row.CouncilArea} | Cantidad de Registros: {row.Cantidad}")
            else:
                print(f"Ciudad: {row.CouncilArea} | Barrio: {row.Suburb} | Cantidad de Registros: {row.Cantidad}")
            print("\n")                
```

# Antes de realizar la unión de datasets, nos aseguraremos que el dataset de Kaggle no posea valores nulos en el campo Postcode

```python
# check for missing values
kaggle_df["Postcode"].isnull().sum()
```

A continuación unimos ambos datasets, utilizando el tipo de unión "LEFT JOIN"

![imagen](./left_join.png)

Si no existe ninguna coincidencia para alguna de las filas de la tabla de la izquierda (kaggle_melbourne), de igual forma todos los resultados de esta se muestran.

```python
# An SQL query is executed and transformed into a dataset
join_df = pd.read_sql("SELECT * FROM kaggle_melbourne LEFT JOIN zip_airbnb_melbourne ON kaggle_melbourne.Postcode = zip_airbnb_melbourne.zipcode", engine.connect())
```
